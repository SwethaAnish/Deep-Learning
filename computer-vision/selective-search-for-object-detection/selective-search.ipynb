```python
#import the libraries 
import selective_search
import skimage
import matplotlib.pyplot as plt
import cv2
```

```python
#Read input image
input_image = cv2.imread("D:/SWETHA/LEARNING/python/leaves.jpg")
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(input_image)
```

```python
#applying selective search algorithm for mode = 'fast'
bounding_boxes_fast = selective_search.selective_search(input_image, mode='fast')
bounding_box_filters = selective_search.box_filter(boxes=bounding_boxes_fast,min_size=4,topN=30)
for x1, y1, x2, y2 in bounding_box_filters:
    cv2.rectangle(input_image,(x1,y1),(x2,y2),(255,0,0),1)
plt.imshow(input_image)
```
